```python
import ires
import numpy as np
```

## Create custom moment functions

```python
moment_funcs = {
    'max': lambda x: np.max(x, axis=-1),
    'min': lambda x: np.min(x, axis=-1),
}
moment_names = {
    'max': "Max",
    'min': "Min",
}
```

```python
i = ires.from_data(np.random.rand(25), win=10, stride=2, moment_funcs=moment_funcs, moment_names=moment_names)
i.vis()
```

## Visualization Settings
