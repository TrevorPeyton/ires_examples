```python
import ires
import numpy as np
```

## Single Dimensional Data
### Load random data and visualize it

```python
i = ires.from_data(np.random.rand(25), win=10, stride=2)
i.vis()
```

### View data

```python
i.moment_data.shape
```

```python
i.moment_data
```

## Multidimensional Data

```python
i = ires.from_data(np.random.rand(10, 25), win=10, stride=2)
i.moment_data.shape
```

### Visualize two different data from first dimension

```python
i.vis(0)
```

```python
i.vis(1)
```

### Higher dimensions

```python
i = ires.from_data(np.random.rand(20, 10, 25), win=10, stride=2)
i.moment_data.shape
```

```python
i.vis((0, 0))
```

```python
i.vis((1, 1))
```
